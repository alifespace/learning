## 1.2 list的常见函数

1. len(): 获得list类型的长度；
2. lst[0]: 访问指定位置的值；
3. lst[-1]: 访问最后一个值；
4. lst[0:3]: 访问一段值，不包括后面的位置；
5. lst5[::3]: 从第一个值开始，每隔2个值取一个值；
6. lst5[::-1]: 逆序列表；
7. first, second, *other = lst5: 解构

````python
print(len(lst5))
print(lst5)
print(lst5[0])
print(lst5[-1])
print(lst5[0:3])
print(lst5[:3])
print(lst5[3:])
print(lst5[::3])
print(lst5[::-1])
print(lst5[::-2])
first, second, *other = lst5
print(first, second, other)
first, *other, last = lst5
print(first, other, last)
````

```python
items = [

```


````python
items = [
    ("product1", 10),
    ("product2", 9),
    ("product3", 12)
]

items.sort()
print(items)

# 这个函数返回用哪个值进行排序
def sort_item(item):
  return item[1]

items.sort(key=sort_item)
print(items)
````

````python
# 使用 lambda 进行排序
items = [
    ("product1", 10),
    ("product2", 9),
    ("product3", 12)
]

items.sort(key=lambda item: item[1])
print(items)
````

````python
# 使用 map 函数

items = [
    ("product1", 10),
    ("product2", 9),
    ("product3", 12)
]

prices = map(lambda item: item[1], items)
print(list(prices))
````
